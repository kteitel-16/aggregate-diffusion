```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Radial-symmetric reaction–diffusion + growth model
@author: Kelly Teitel
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.colors import ListedColormap   # <-- ADDED
import pandas as pd
import os
from datetime import datetime

# ===========================================
# ------------ PARAMETERS -------------------
# ===========================================

GRID_SIZE = 400
DT = 0.05  # diffusion time step, 3 minutes per step
MAX_FRAMES = 200

# Diffusion constants
D_PRODUCT = 5.0
D_SUBSTRATE = 5.0

# Radial geometry
RED_RADIUS = 30
GREEN_RADIUS = 60

# Initial concentrations
SUBSTRATE_INIT = 100.0
PRODUCT_INIT = 0.0

# Enzyme kinetics
KM = 0.05
ENZYME_INIT = 1.0
MAX_ENZYME_ACTIVITY = 0.1

# Growth kinetics
GROWTH_RATE_PER_HOUR = np.log(2)
GROWTH_KM = 0.1

# ===========================================
# ------------ RADIAL GRID ------------------
# ===========================================

center = GRID_SIZE // 2
x = np.arange(GRID_SIZE)
y = np.arange(GRID_SIZE)
X, Y = np.meshgrid(x, y, indexing="ij")
R = np.sqrt((X - center)**2 + (Y - center)**2)

# ===========================================
# ------------ HELPER FUNCTIONS -------------
# ===========================================

def radial_average(field, R, dr=1.0):
    r_max = int(R.max())
    bins = np.arange(0, r_max + dr, dr)
    means = np.zeros(len(bins) - 1)

    for k in range(len(bins) - 1):
        mask = (R >= bins[k]) & (R < bins[k + 1])
        means[k] = field[mask].mean() if np.any(mask) else np.nan

    return 0.5 * (bins[:-1] + bins[1:]), means


def diffusion_2d(field, D, dt):
    lap = (
        np.roll(field, 1, axis=0) + np.roll(field, -1, axis=0) +
        np.roll(field, 1, axis=1) + np.roll(field, -1, axis=1) -
        4 * field
    )
    return field + D * dt * lap

# ===========================================
# ------------ INITIALIZATION ----------------
# ===========================================

timestamp = datetime.now().strftime("run_%Y-%m-%d_%H-%M-%S")
output_dir = os.path.join("simulation_outputs", timestamp)
os.makedirs(output_dir, exist_ok=True)

grid = np.zeros((GRID_SIZE, GRID_SIZE), dtype=int)
substrate = np.zeros((GRID_SIZE, GRID_SIZE), dtype=float)
product = np.full((GRID_SIZE, GRID_SIZE), PRODUCT_INIT, dtype=float)
enzyme = np.zeros((GRID_SIZE, GRID_SIZE), dtype=float)

red_cells = R <= RED_RADIUS
green_cells = (R > RED_RADIUS) & (R <= GREEN_RADIUS)

grid[red_cells] = 1
grid[green_cells] = 2
enzyme[green_cells] = ENZYME_INIT

substrate[R > GREEN_RADIUS] = SUBSTRATE_INIT

substrate_frames = []
product_frames = []
cell_frames = []   # <-- ADDED

# ===========================================
# ------------ VISUALIZATION ----------------
# ===========================================

fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # <-- CHANGED

axes[0].set_title("Substrate")
axes[1].set_title("Product")
axes[2].set_title("Cells")

im_sub = axes[0].imshow(substrate, cmap="Blues", vmin=0, vmax=SUBSTRATE_INIT)
im_prod = axes[1].imshow(product, cmap="Reds", vmin=0, vmax=1.0)

cell_cmap = ListedColormap(["black", "red", "green"])  # <-- ADDED
im_cells = axes[2].imshow(grid, cmap=cell_cmap, vmin=0, vmax=2)  # <-- ADDED

for ax in axes:
    ax.axis("off")

# ===========================================
# ------------ SIMULATION LOOP ---------------
# ===========================================

for frame in range(MAX_FRAMES):

    vmax = enzyme * MAX_ENZYME_ACTIVITY
    rate = vmax * substrate / (KM + substrate)
    substrate -= rate * DT
    product += rate * DT
    substrate = np.clip(substrate, 0, None)

    substrate = diffusion_2d(substrate, D_SUBSTRATE, DT)
    product   = diffusion_2d(product, D_PRODUCT, DT)

    substrate = np.clip(substrate, 0, None)

    new_grid = grid.copy()
    new_enzyme = enzyme.copy()

    for cell_type in [1, 2]:
        positions = np.argwhere(grid == cell_type)
        np.random.shuffle(positions)

        for i, j in positions:
            p = product[i, j]
            grow_prob = GROWTH_RATE_PER_HOUR * DT * p / (GROWTH_KM + p)

            if np.random.rand() < grow_prob:
                nbrs = [(i + di, j + dj)
                        for di in [-1, 0, 1]
                        for dj in [-1, 0, 1]
                        if not (di == 0 and dj == 0)]

                nbrs = [(ni % GRID_SIZE, nj % GRID_SIZE) for ni, nj in nbrs]
                empty = [n for n in nbrs if grid[n] == 0]

                if empty:
                    ni, nj = empty[np.random.randint(len(empty))]
                    new_grid[ni, nj] = cell_type

                    if cell_type == 2:
                        parent = min(new_enzyme[i, j] + 0.01, 2.0)
                        new_enzyme[i, j] = parent
                        new_enzyme[ni, nj] = parent

    grid = new_grid
    enzyme = new_enzyme

    r_vals, prod_r = radial_average(product, R)
    pd.DataFrame({"r": r_vals, "product": prod_r}).to_csv(
        f"{output_dir}/radial_product_{frame:04d}.csv", index=False
    )

    substrate_frames.append(substrate.copy())
    product_frames.append(product.copy())
    cell_frames.append(grid.copy())   # <-- ADDED

    print(f"Frame {frame:04d} | Cell density = {np.mean(grid > 0):.3f}")

# ===========================================
# ------------ SAVE VIDEO -------------------
# ===========================================

def update(frame):
    im_sub.set_data(substrate_frames[frame])
    im_prod.set_data(product_frames[frame])
    im_cells.set_data(cell_frames[frame])   # <-- ADDED
    return im_sub, im_prod, im_cells

ani = FuncAnimation(
    fig,
    update,
    frames=len(substrate_frames),
    blit=True
)

ani.save(
    os.path.join(output_dir, "substrate_product_cells.gif"),
    writer="pillow",
    fps=10
)

plt.close()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Radial-symmetric reaction–diffusion + growth model
Red and green cells saved as separate animations

@author: Kelly Teitel
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.colors import ListedColormap
import pandas as pd
import os
from datetime import datetime

# ===========================================
# ------------ PARAMETERS -------------------
# ===========================================

GRID_SIZE = 400
DT = 0.05
MAX_FRAMES = 200

# Diffusion constants
D_PRODUCT = 5.0
D_SUBSTRATE = 5.0

# Radial geometry
RED_RADIUS = 30
GREEN_RADIUS = 60

# Initial concentrations
SUBSTRATE_INIT = 100.0
PRODUCT_INIT = 0.0

# Enzyme kinetics
KM = 0.05
ENZYME_INIT = 1.0
MAX_ENZYME_ACTIVITY = 0.1

# Growth kinetics
GROWTH_RATE_PER_HOUR = np.log(2)
GROWTH_KM = 0.1

# ===========================================
# ------------ RADIAL GRID ------------------
# ===========================================

center = GRID_SIZE // 2
x = np.arange(GRID_SIZE)
y = np.arange(GRID_SIZE)
X, Y = np.meshgrid(x, y, indexing="ij")
R = np.sqrt((X - center)**2 + (Y - center)**2)

# ===========================================
# ------------ HELPER FUNCTIONS -------------
# ===========================================

def radial_average(field, R, dr=1.0):
    r_max = int(R.max())
    bins = np.arange(0, r_max + dr, dr)
    means = np.zeros(len(bins) - 1)

    for k in range(len(bins) - 1):
        mask = (R >= bins[k]) & (R < bins[k + 1])
        means[k] = field[mask].mean() if np.any(mask) else np.nan

    return 0.5 * (bins[:-1] + bins[1:]), means


def diffusion_2d(field, D, dt):
    lap = (
        np.roll(field, 1, axis=0) + np.roll(field, -1, axis=0) +
        np.roll(field, 1, axis=1) + np.roll(field, -1, axis=1) -
        4 * field
    )
    return field + D * dt * lap

# ===========================================
# ------------ INITIALIZATION ----------------
# ===========================================

timestamp = datetime.now().strftime("run_%Y-%m-%d_%H-%M-%S")
output_dir = os.path.join("simulation_outputs", timestamp)
os.makedirs(output_dir, exist_ok=True)

grid = np.zeros((GRID_SIZE, GRID_SIZE), dtype=int)
substrate = np.zeros((GRID_SIZE, GRID_SIZE), dtype=float)
product = np.full((GRID_SIZE, GRID_SIZE), PRODUCT_INIT, dtype=float)
enzyme = np.zeros((GRID_SIZE, GRID_SIZE), dtype=float)

red_cells = R <= RED_RADIUS
green_cells = (R > RED_RADIUS) & (R <= GREEN_RADIUS)

grid[red_cells] = 1
grid[green_cells] = 2
enzyme[green_cells] = ENZYME_INIT

substrate[R > GREEN_RADIUS] = SUBSTRATE_INIT

substrate_frames = []
product_frames = []
red_cell_frames = []
green_cell_frames = []

# ===========================================
# ------------ SIMULATION LOOP ---------------
# ===========================================

for frame in range(MAX_FRAMES):

    vmax = enzyme * MAX_ENZYME_ACTIVITY
    rate = vmax * substrate / (KM + substrate)
    substrate -= rate * DT
    product += rate * DT
    substrate = np.clip(substrate, 0, None)

    substrate = diffusion_2d(substrate, D_SUBSTRATE, DT)
    product   = diffusion_2d(product, D_PRODUCT, DT)
    substrate = np.clip(substrate, 0, None)

    new_grid = grid.copy()
    new_enzyme = enzyme.copy()

    for cell_type in [1, 2]:
        positions = np.argwhere(grid == cell_type)
        np.random.shuffle(positions)

        for i, j in positions:
            p = product[i, j]
            grow_prob = GROWTH_RATE_PER_HOUR * DT * p / (GROWTH_KM + p)

            if np.random.rand() < grow_prob:
                nbrs = [(i + di, j + dj)
                        for di in [-1, 0, 1]
                        for dj in [-1, 0, 1]
                        if not (di == 0 and dj == 0)]

                nbrs = [(ni % GRID_SIZE, nj % GRID_SIZE) for ni, nj in nbrs]
                empty = [n for n in nbrs if grid[n] == 0]

                if empty:
                    ni, nj = empty[np.random.randint(len(empty))]
                    new_grid[ni, nj] = cell_type

                    if cell_type == 2:
                        parent = min(new_enzyme[i, j] + 0.01, 2.0)
                        new_enzyme[i, j] = parent
                        new_enzyme[ni, nj] = parent

    grid = new_grid
    enzyme = new_enzyme

    r_vals, prod_r = radial_average(product, R)
    pd.DataFrame({"r": r_vals, "product": prod_r}).to_csv(
        f"{output_dir}/radial_product_{frame:04d}.csv", index=False
    )

    substrate_frames.append(substrate.copy())
    product_frames.append(product.copy())
    red_cell_frames.append((grid == 1).astype(int))
    green_cell_frames.append((grid == 2).astype(int))

    print(f"Frame {frame:04d} | Cell density = {np.mean(grid > 0):.3f}")

# ===========================================
# ------------ SUBSTRATE + PRODUCT GIF -------
# ===========================================

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].set_title("Substrate")
axes[1].set_title("Product")

im_sub = axes[0].imshow(substrate_frames[0], cmap="Blues",
                        vmin=0, vmax=SUBSTRATE_INIT)
im_prod = axes[1].imshow(product_frames[0], cmap="Reds",
                         vmin=0, vmax=1.0)

for ax in axes:
    ax.axis("off")

def update_main(frame):
    im_sub.set_data(substrate_frames[frame])
    im_prod.set_data(product_frames[frame])
    return im_sub, im_prod

ani_main = FuncAnimation(
    fig,
    update_main,
    frames=len(substrate_frames),
    blit=True
)

ani_main.save(
    os.path.join(output_dir, "substrate_product.gif"),
    writer="pillow",
    fps=10
)

plt.close(fig)

# ===========================================
# ------------ RED CELLS GIF -----------------
# ===========================================

fig_red, ax_red = plt.subplots(figsize=(5, 5))
ax_red.set_title("Red Cells")

im_red = ax_red.imshow(
    red_cell_frames[0],
    cmap=ListedColormap(["black", "red"]),
    vmin=0,
    vmax=1
)
ax_red.axis("off")

def update_red(frame):
    im_red.set_data(red_cell_frames[frame])
    return (im_red,)

ani_red = FuncAnimation(
    fig_red,
    update_red,
    frames=len(red_cell_frames),
    blit=True
)

ani_red.save(
    os.path.join(output_dir, "red_cells.gif"),
    writer="pillow",
    fps=10
)

plt.close(fig_red)

# ===========================================
# ------------ GREEN CELLS GIF ---------------
# ===========================================

fig_green, ax_green = plt.subplots(figsize=(5, 5))
ax_green.set_title("Green Cells")

im_green = ax_green.imshow(
    green_cell_frames[0],
    cmap=ListedColormap(["black", "green"]),
    vmin=0,
    vmax=1
)
ax_green.axis("off")

def update_green(frame):
    im_green.set_data(green_cell_frames[frame])
    return (im_green,)

ani_green = FuncAnimation(
    fig_green,
    update_green,
    frames=len(green_cell_frames),
    blit=True
)

ani_green.save(
    os.path.join(output_dir, "green_cells.gif"),
    writer="pillow",
    fps=10
)

plt.close(fig_green)
```
